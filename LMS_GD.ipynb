# LMS with gradient descent

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# choose 50 point randomly (uniform distribution) from [0, 10)
x = 10*np.random.rand(50,1)

# Let y = x-3 with random perturbation (normal distribution)
y = x - 3 + 0.2*np.random.randn(50,1)

# scatter plot
plt.scatter(x,y,s=100)
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20,rotation=0)
```

```python
# learning rate
learning_rate = 0.001

# number of iteration
n_iterations = 10000

# parameter initialization, choose randomly from normal distribution
theta = np.random.randn(2,1)

# define the xx by adding ones
xx = np.hstack((np.ones((50,1)), x))

# gradient descent
for iteration in range(n_iterations):

    # h(x) = theta^T*x
    prediction = np.dot(xx,theta)

    # prediction_error
    prediction_error = y - prediction
    mse = (prediction_error**2).mean()

    # x的shape(50,2) output_error的shapeｊ為(50,1) gradients為(2,1)
    gradients = np.dot(xx.T,prediction_error)/50
    
    # update the parameters
    theta += learning_rate*gradients
    plt.plot(iteration,mse,'o-',color='red')

plt.xlabel('iteration',fontsize=20)
plt.ylabel('mse',fontsize=20)
plt.show()

print('mse= ', mse)
print('theta_0= ',theta[0])
print('theta_1= ',theta[1])
```

```python
xs = np.linspace(0, 10, 10)
ys = theta[0] + theta[1]*xs
plt.plot(x, y, 'o',markersize=15,alpha=0.3)
plt.plot(xs,ys,linewidth=5)
```
